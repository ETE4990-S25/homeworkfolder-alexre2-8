````python
class animals (object):
    """Create the base attributes of an animal"""
    def __init__(self, name, age):
        """Initialize name, age of animal."""
        self.name = name
        self.age = age
        #self.species = species

class livestockAnimal (animals):
    """Establish livestock category of animal"""
    def __init__(self, name, age):
        """Initialize name, age of animal."""
        animals.__init__(self, name, age)
        self.suppMeat=False
        self.suppWork=False
        self.suppMats=False
        self.suppDairy=False
        self.suppEggs=False


class Pig(livestockAnimal):
    def __init__(self, name, age,suppMeat):
        """Initialize name, age of animal."""
        livestockAnimal.__init__(self, name, age)
        if suppMeat==True:
            self.Meattype="Pork"
class Sheep(livestockAnimal):
    def __init__(self, name, age,suppMeat,suppMats):
        """Initialize name, age of animal."""
        livestockAnimal.__init__(self, name, age)
        if suppMeat==True:
            self.Meattype="Mutton"
        if suppMats==True:
            self.Matstype="Wool"
class Cow(livestockAnimal):
    def __init__(self, name, age,suppMeat,suppWork,suppMats,suppDairy):
        """Initialize name, age of animal."""
        livestockAnimal.__init__(self, name, age)
        if suppMeat==True:
            self.Meattype="Beef"
        if suppWork==True:
            self.Worktype="Plowing"
        if suppMats==True:
            self.Matstype="Leather"
        if suppDairy==True:
            self.Dairyprod="Milk, Cheese"
class Chicken(livestockAnimal):
    def __init__(self, name, age,suppMeat,suppMats,suppEggs):
        """Initialize name, age of animal."""
        livestockAnimal.__init__(self, name, age)
        if suppMeat==True:
            self.Meattype="Poultry"
        if suppEggs==True:
            self.EggType="Chicken Eggs"
        if suppMats==True:
            self.Matstype="Feathers"
class Goat(livestockAnimal):
    def __init__(self, name, age,suppMeat,suppWork,suppMats,suppDairy):
        """Initialize name, age of animal."""
        livestockAnimal.__init__(self, name, age)
        if suppMeat==True:
            self.Meattype="Mutton"
        if suppWork==True:
            self.Worktype="Plowing,Clearing"
        if suppMats==True:
            self.Matstype="Wool"
        if suppDairy==True:
            self.Dairyprod="Milk, Cheese"  

def get_bool(question):
    response = input(question)
    while response not in ("True","true","False","false"):
        response = input("Invalid input, please try again")
    if response in ("True","true"):
        return True
    
def animCreate():
    animSpecies=input("Please enter an animal type:  ")
    if animSpecies not in ("Cow","Sheep","Pig","Goat","Chicken"):
        return "Invalid Animal Type, please try again"
    
    animName=input("Please enter an animal name:  ")
    animAge=int(input("Please input animal Age (numerically in years):"))
    animMeat=get_bool("Will this animal provide meat? True or False: ")
    if animSpecies!= "Pig":
        animMats=get_bool("Will this animal produce Material?")
        if animSpecies== "Chicken":
            animEggs=get_bool("Are you harvesting eggs? T or F")
        elif animSpecies== "Goat" or "Cow":
            animDairy=get_bool("Will this animal be providing dairy? T or F")
            animWork=get_bool("Will this animal be providing labor? T or F")
    if animSpecies=="Pig":
        finalAnim=Pig(animName,animAge,animMeat)
        
    elif animSpecies=="Cow":
        finalAnim=Cow(animName,animAge,animMeat,animWork,animMats,animDairy)
    elif animSpecies=="Chicken":
        finalAnim=Chicken(animName,animAge,animMeat,animMats,animEggs)
    elif animSpecies=="Sheep":
        finalAnim=Sheep(animName,animAge,animMeat,animMats)
    elif animSpecies=="Goat":
        finalAnim=Goat(animName,animAge,animMeat,animWork,animMats,animDairy)
    return finalAnim


farmlist=[]
while True:
    if input("Would you like to add an animal to the farm?")=="yes":
        farmanim=animCreate()
        farmlist.append(farmanim)
        
    else:
        break
print(farmlist)

````
